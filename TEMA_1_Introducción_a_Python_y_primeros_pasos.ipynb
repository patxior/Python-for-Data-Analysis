## **Slicing**

La t√©cnica de slicing (rebanado o segmentado en espa√±ol) es una forma de obtener una subcadena (substring) o secci√≥n de una cadena (string). Se utiliza para extraer una parte espec√≠fica de una cadena.

La sintaxis del slicing es la siguiente:
```
cadena[inicio:fin:paso]
```
* `cadena`: La cadena de la cual se quiere obtener una subcadena.
* `inicio`: El √≠ndice (posici√≥n) de la cadena desde donde se iniciar√° el slice. **Si no se especifica, se tomar√° el valor 0.**
* `fin`: El √≠ndice de la cadena donde se detendr√° el slice. **Si no se especifica, se tomar√° el valor `len(cadena)`**.
* `paso`: El n√∫mero de caracteres que se saltar√° en cada iteraci√≥n del slice. **Si no se especifica, se tomar√° el valor 1**.

Es importante tener en cuenta que el slice incluir√° el car√°cter en la posici√≥n inicio pero no incluir√° el car√°cter en la posici√≥n fin.


```
+---+---+---+---+---+---+\
 | P | y | t | h | o | n |\
 +---+---+---+---+---+---+\
 0   1   2   3   4   5   6\
-6  -5  -4  -3  -2  -1\
```



````python
cadena = "Hola Mundo"
````

Obtener la subcadena desde el √≠ndice 0 hasta el √≠ndice 3 (no incluye el 3)

````python
subcadena = cadena[:3] #cadena[inicio:fin:paso] 

"""
Esto es equivalente a:
subcadena = cadena[0:3:1]

Pero como inicio es 0 y paso es 1, se pueden omitir
"""

print(subcadena) 
````

Obtener la subcadena desde el √≠ndice 3 hasta el final

````python
subcadena = cadena[3:]
print(subcadena) 

# Esto es equivalente a:
# subcadena = cadena[3:len(cadena):1]
# Pero como fin es len(cadena) y paso es 1, se pueden omitir
````

Obtener la subcadena saltando de 2 en 2 caracteres

````python
subcadena = cadena[::2]
print(subcadena) 

# Esto es equivalente a:
# subcadena = cadena[0:len(cadena):2]
# Pero como inicio es 0 y fin es len(cadena), se pueden omitir
````

Obtener la subcadena invertida

````python
subcadena = cadena[::-1]
print(subcadena) 
````

# **Booleanos**

### **comparadores**
Los comparadores en Python son operadores que permiten comparar dos valores y determinar si una relaci√≥n entre ellos es verdadera o falsa. Los principales comparadores en Python son:
* `==`: igual a
* `!=`: distinto de
* `>`: mayor que
* `<`: menor que
* `>=`: mayor o igual que
* `<=`: menor o igual que

````python
x = 5
y = 7

print(x == y)  # False
print(x != y)  # True
print(x > y)   # False
print(x < y)   # True
print(x >= y)  # False
print(x <= y)  # True
````

### **operadores l√≥gicos**
Son utilizados para evaluar expresiones booleanas.
* `and`: devuelve `True` si ambos operandos son `True`. De lo contrario, devuelve `False`.
* `or`: devuelve `True` si al menos uno de los operandos es True. De lo contrario, devuelve `False`.
* `not`: devuelve el inverso booleano del operando. Si el operando es `True`, devuelve `False`. Si el operando es False, devuelve `True`.

````python
x = 5
y = 10
z = 15

# operador and
print(x < y and y < z)  # True
print(x < y and z < y)  # False

# operador or
print(x > y or y > z)   # False
print(x < y or z < y)   # True

# operador not
print(not(x < y))       # False
print(not(x > y))       # True
````

### **operador `in`**
El operador `in` se utiliza para verificar si un elemento est√° presente en una secuencia, como una cadena de caracteres, una lista, una tupla, un conjunto o un diccionario.

Devuelve `True` si el elemento se encuentra en la secuencia y `False` en caso contrario.
```
elemento in secuencia
```
* `elemento` es el valor que se busca en la secuencia.
* `secuencia` es la secuencia en la que se busca.




````python
elemento  = 'Hola'
secuencia =  'Hola Mundo'
print(elemento in secuencia)
````
