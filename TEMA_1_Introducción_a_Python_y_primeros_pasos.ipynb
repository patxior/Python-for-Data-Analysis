## **Slicing**

La técnica de slicing (rebanado o segmentado en español) es una forma de obtener una subcadena (substring) o sección de una cadena (string). Se utiliza para extraer una parte específica de una cadena.

La sintaxis del slicing es la siguiente:
```
cadena[inicio:fin:paso]
```
* `cadena`: La cadena de la cual se quiere obtener una subcadena.
* `inicio`: El índice (posición) de la cadena desde donde se iniciará el slice. **Si no se especifica, se tomará el valor 0.**
* `fin`: El índice de la cadena donde se detendrá el slice. **Si no se especifica, se tomará el valor `len(cadena)`**.
* `paso`: El número de caracteres que se saltará en cada iteración del slice. **Si no se especifica, se tomará el valor 1**.

Es importante tener en cuenta que el slice incluirá el carácter en la posición inicio pero no incluirá el carácter en la posición fin.


```
+---+---+---+---+---+---+\
 | P | y | t | h | o | n |\
 +---+---+---+---+---+---+\
 0   1   2   3   4   5   6\
-6  -5  -4  -3  -2  -1\
```



````python
cadena = "Hola Mundo"
````

Obtener la subcadena desde el índice 0 hasta el índice 3 (no incluye el 3)

````python
subcadena = cadena[:3] #cadena[inicio:fin:paso] 

"""
Esto es equivalente a:
subcadena = cadena[0:3:1]

Pero como inicio es 0 y paso es 1, se pueden omitir
"""

print(subcadena) 
````

Obtener la subcadena desde el índice 3 hasta el final

````python
subcadena = cadena[3:]
print(subcadena) 

# Esto es equivalente a:
# subcadena = cadena[3:len(cadena):1]
# Pero como fin es len(cadena) y paso es 1, se pueden omitir
````

Obtener la subcadena saltando de 2 en 2 caracteres

````python
subcadena = cadena[::2]
print(subcadena) 

# Esto es equivalente a:
# subcadena = cadena[0:len(cadena):2]
# Pero como inicio es 0 y fin es len(cadena), se pueden omitir
````

Obtener la subcadena invertida

````python
subcadena = cadena[::-1]
print(subcadena) 
````

# **Booleanos**
Los booleanos son un tipo de dato que solo puede tener dos valores: `True` o `False`. Estos valores son útiles para realizar operaciones de comparación y lógica en programas.

### **comparadores**
Los comparadores en Python son operadores que permiten comparar dos valores y determinar si una relación entre ellos es verdadera o falsa. Los principales comparadores en Python son:
* `==`: igual a
* `!=`: distinto de
* `>`: mayor que
* `<`: menor que
* `>=`: mayor o igual que
* `<=`: menor o igual que

````python
x = 5
y = 7

print(x == y)  # False
print(x != y)  # True
print(x > y)   # False
print(x < y)   # True
print(x >= y)  # False
print(x <= y)  # True
````

### **operadores lógicos**
Son utilizados para evaluar expresiones booleanas.
* `and`: devuelve `True` si ambos operandos son `True`. De lo contrario, devuelve `False`.
* `or`: devuelve `True` si al menos uno de los operandos es True. De lo contrario, devuelve `False`.
* `not`: devuelve el inverso booleano del operando. Si el operando es `True`, devuelve `False`. Si el operando es False, devuelve `True`.

````python
x = 5
y = 10
z = 15

# operador and
print(x < y and y < z)  # True
print(x < y and z < y)  # False

# operador or
print(x > y or y > z)   # False
print(x < y or z < y)   # True

# operador not
print(not(x < y))       # False
print(not(x > y))       # True
````

### **operador `in`**
El operador `in` se utiliza para verificar si un elemento está presente en una secuencia, como una cadena de caracteres, una lista, una tupla, un conjunto o un diccionario.

Devuelve `True` si el elemento se encuentra en la secuencia y `False` en caso contrario.
```
elemento in secuencia
```
* `elemento` es el valor que se busca en la secuencia.
* `secuencia` es la secuencia en la que se busca.




````python
elemento  = 'Hola'
secuencia =  'Hola Mundo'
print(elemento in secuencia)
````

# **Listas**
Las listas nos permiten almacenar una colección ordenada de elementos. Cada elemento en una lista puede ser de cualquier tipo de datos, como enteros, flotantes, cadenas de caracteres, entre otros.

Para crear una lista en Python, utilizamos corchetes `[]` y separamos cada elemento con una coma. Por ejemplo, la siguiente línea crea una lista de enteros.

````python
numeros = [1, 2, 3, 4, 5]
print(numeros)
print(type(numeros))
````

Podemos acceder a los elementos de una lista utilizando su índice, que es un número que representa la posición del elemento en la lista. **El primer elemento de la lista tiene un índice de 0**, el segundo tiene un índice de 1, y así sucesivamente. Por ejemplo, para acceder al segundo elemento de la lista numeros, podemos utilizar.

````python
primer_numero = numeros[0]
primer_numero
````

````python
segundo_numero = numeros[1]
segundo_numero
````

````python
ultimo_numero = numeros[-1]
ultimo_numero
````

# **Built in functions (Funciones integradas)**
Las funciones integradas (también conocidas como funciones predefinidas) son un conjunto de funciones que se proporcionan con el lenguaje y que se pueden utilizar sin necesidad de definirlas previamente. Algunas de las funciones integradas más comunes en Python son:

* `print()`: Imprime un objeto en la consola.
* `len()`: Devuelve la longitud de un objeto iterable.
* `input()`: Permite al usuario ingresar un valor por teclado.
* `type()`: Devuelve el tipo de un objeto.
* `int()`: Convierte un objeto a un entero.
* `float()`: Convierte un objeto a un número de punto flotante.
* `str()`: Convierte un objeto a una cadena de caracteres.
* `list()`: Convierte un objeto a una lista.
* `dict()`: Convierte un objeto a un diccionario.
* `tuple()`: Convierte un objeto a una tupla.
