## **Seguimos con las listas, `list`**

### **Métodos que tiene `list`**
* `append(elemento)`: Agrega un elemento al final de la lista.
* `extend(iterable)`: Agrega los elementos de otro iterable (lista, tupla, conjunto, etc.) al final de la lista.
* `insert(posición, elemento)`: Inserta un elemento en una posición específica en la lista.
* `remove(elemento)`: Elimina la primera ocurrencia del elemento especificado de la lista.
* `pop([índice])`: Elimina y devuelve el elemento en la posición especificada. Si no se especifica el índice, se elimina y devuelve el último elemento de la lista.
* `index(elemento)`: Devuelve el índice de la primera ocurrencia del elemento especificado en la lista.
* `count(elemento)`: Devuelve el número de veces que el elemento especificado aparece en la lista.
* `sort()`: Ordena los elementos de la lista en orden ascendente.
* `reverse()`: Invierte el orden de los elementos en la lista.
* `clear()`: Elimina todos los elementos de una lista.


````python
lista = [1, 2, 3]
lista.append(4)
print(lista)  # [1, 2, 3, 4]
````

````python
lista1 = [1, 2, 3]
lista2 = [4, 5, 6]
lista1.extend(lista2)
print(lista1)  # [1, 2, 3, 4, 5, 6]
````

````python
lista = [1, 2, 3]
lista.insert(1, 4)
print(lista)  # [1, 4, 2, 3]
````

````python
lista = [1, 2, 3, 2]
lista.remove(2)
print(lista)  # [1, 3, 2]
````

````python
lista = [1, 2, 3, 2]
índice = lista.index(2)
print(índice)  # 1
````

````python
lista = [1, 2, 3, 2]
n = lista.count(2)
print(n)  # 2
````

````python
lista = [1, 2, 3]
elemento = lista.pop(1)
print(lista)  # [1, 3]
print(elemento)  # 2
````

````python
lista = [1, 2, 3]
lista.reverse()
print(lista)  # [3, 2, 1]
````

````python
lista = [1, 2, 3, 4, 5]
lista.clear()
print(lista) # Output: []
````

### **List Comprehensions**
 
Es una forma concisa y poderosa de crear una nueva lista a partir de otra lista o iterable. Se basa en la construcción de una lista utilizando una sintaxis compacta y legible en una sola línea de código.

```
nueva_lista = [expresión for elemento in iterable if condición]
```

* `expresión` es el valor que queremos que tenga cada `elemento` en la nueva lista
* `elemento` es la variable que representa cada elemento en el iterable original
* `iterable` es la lista, tupla, set, u otro iterable del que se tomarán los elementos
* `condición` es una expresión booleana opcional que filtra los elementos del iterable original que se incluyen en la nueva lista

Puede ser una herramienta muy útil para reducir la cantidad de código necesario para crear nuevas listas, y para hacer que el código sea más fácil de leer y entender.


````python
# Crear una lista de números pares del 0 al 10:
pares = [x for x in range(11) if x % 2 == 0]
print(pares)  # output: [0, 2, 4, 6, 8, 10]
````

````python
# Crear una lista de las letras de una cadena de texto:
texto = "Hola Mundo"
letras = [letra for letra in texto]
print(letras)  # output: ['H', 'o', 'l', 'a', ' ', 'M', 'u', 'n', 'd', 'o']
````

````python
# Crear una lista de los cuadrados de los números del 1 al 5:
cuadrados = [x**2 for x in range(1, 6)]
print(cuadrados)  # output: [1, 4, 9, 16, 25]
````

````python
# Crear una lista de los valores de una lista multiplicados por 2:
lista = [1, 2, 3, 4, 5]
multiplicados = [x*2 for x in lista]
print(multiplicados)  # output: [2, 4, 6, 8, 10]
````

### **Nested Lists**
Una listas anidada (Nested List) es una lista que contiene otras listas dentro de ella. Es decir, cada elemento de la lista principal es a su vez otra lista.




Por ejemplo, si queremos crear una lista de listas para representar una matriz de 3x3, podríamos hacerlo de la siguiente manera:

````python
mat = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9],
]
print(mat)
````

En este caso, matriz es una lista que contiene tres listas dentro de ella. Cada una de estas listas representa una fila de la matriz.

Podemos acceder a los elementos de una lista anidada utilizando dos índices: uno para acceder a la lista interna y otro para acceder al elemento de esa lista.

````python
# Si queremos acceder al elemento en la segunda fila y tercera columna
elemento = mat[1][2]
print(elemento)  # 6
````

Aquí, `mat[1]` devuelve la segunda lista interna (que representa la segunda fila de la matriz), y `mat[1][2]` devuelve el tercer elemento de esa lista (que representa la tercera columna de la matriz).

## **Funcion `del`**

La función `del` elimina objetos de la memoria. Puede eliminar una variable, un índice de una lista o elementos dentro de un diccionario. La sintaxis es simplemente del objeto, donde objeto puede ser una variable, índice o elemento.

Por ejemplo, para eliminar una variable `x`, se puede utilizar `del x`. Para eliminar un elemento de una lista, se puede utilizar `del lista[indice]` donde lista es la lista y indice es el índice del elemento que se desea eliminar.

## **Tuplas y Sets**
Las **tuplas** son otrotipo de contenedor. Son **inmutables**, lo que significa que no se pueden cambiar una vez creadas, mientras que las listas son mutables y se pueden cambiar.

Las tuplas (como las listas) se pueden indexar y rebanar para acceder a elementos individuales o a subconjuntos de elementos.

Las tuplas se crean colocando los objetos dentro de paréntesis y separándolos por comas. Por ejemplo:

```
tupla = (1, 2, 3)
```


Un **set** (conjunto) es una colección desordenada de elementos únicos. Se puede crear un conjunto utilizando la función incorporada `set()` o utilizando llaves `{}` con elementos separados por comas.

Por ejemplo:

```
my_set = set()
my_set.add(1)
my_set.add(2)
my_set.add(3)
print(my_set) # Salida: {1, 2, 3}

my_set = {1, 2, 3}
print(my_set) # Salida: {1, 2, 3}
```


Los conjuntos son útiles cuando se quiere eliminar elementos duplicados de una lista o cuando se quiere verificar si un elemento está presente en una colección de elementos de forma rápida y eficiente. También se pueden realizar operaciones de conjunto como unión, intersección y diferencia utilizando los métodos incorporados `union()`, `intersection()` y `difference()`, respectivamente.


*   El método **`union()`** devuelve un conjunto que contiene todos los elementos de ambos conjuntos, eliminando duplicados. Se puede utilizar el operador `|` para realizar la misma operación.
*   El método **`intersection()`** devuelve un conjunto que contiene los elementos que están presentes en ambos conjuntos. Se puede utilizar el operador `&` para realizar la misma operación.
*   El método **`difference()`** devuelve un conjunto que contiene los elementos que están en el primer conjunto pero no en el segundo conjunto. Se puede utilizar el operador `-` para realizar la misma operación.



````python
# Definir conjuntos
set1 = {1, 2, 3}
set2 = {2, 3, 4}

# union
set3 = set1.union(set2)
set4 = set1 | set2
print(set3)  # {1, 2, 3, 4}
print(set4)  # {1, 2, 3, 4}

# intersection
set5 = set1.intersection(set2)
set6 = set1 & set2
print(set5)  # {2, 3}
print(set6)  # {2, 3}

# difference
set7 = set1.difference(set2)
set8 = set1 - set2
print(set7)  # {1}
print(set8)  # {1}
````

## **Diccionarios**

Es una colección **no ordenada** de elementos que se almacenan en pares **clave-valor (key-value)**. Cada elemento del diccionario es una pareja clave-valor, donde la clave es una cadena o un número que identifica al valor. Los diccionarios se definen usando llaves `{}` y los pares clave-valor se separan por dos puntos `:`, y los elementos se separan por comas `,`

````python
edades = {'Aitor': 28, 'Irati': 31, 'Ane': 29}
````

En este caso, las claves son los nombres y los valores son sus respectivas edades.

Para acceder a un valor en un diccionario, se utiliza su clave entre corchetes `[]`


````python
print(edades['Irati']) # Imprime 31
````

Si se intenta acceder a una clave que no está en el diccionario, se producirá un error `KeyError`. Para evitar este error se puede usar el método `get()` que devuelve el valor correspondiente a la clave, o un valor por defecto si la clave no existe en el diccionario.

````python
# edades['Ander']  # KeyError
````

````python
print(edades.get('Ane'))  # 29
````

````python
print(edades.get('Ander'))  # 29
````

Para agregar un nuevo elemento al diccionario se puede asignar un valor a una nueva clave:

```
edades['Ander'] = 21
```

Para modificar el valor de un elemento existente, se puede asignar un nuevo valor a la clave correspondiente:

```
edades['Ander'] = 37
```

Para eliminar un elemento del diccionario se utiliza la instrucción del seguida de la clave:

```
del edades['Ander']
```




````python
edades['Ander'] = 21
edades
````

````python
edades['Ander'] = 37
edades
````

````python
del edades['Ander']
edades
````

Los diccionarios también tienen varios métodos útiles, como `keys()`, que devuelve una lista de las claves del diccionario, `values()`, que devuelve una lista de los valores del diccionario, y `items()`, que devuelve una lista de tuplas que contienen cada pareja clave-valor.

````python
print(edades.values())
````

````python
print(edades.keys())
````

````python
print(edades.items())
````
